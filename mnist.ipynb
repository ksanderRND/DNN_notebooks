```python
import numpy
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from keras.datasets import mnist
```

```python
#Set a seed for repeatability of results.
numpy.random.seed(42)

# load dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
#image size conversion (60 thousand images consist of 784 pixels)
X_train=X_train.reshape(60000, 784)
X_test=X_test.reshape(10000, 784)

#normalization
X_train=X_train.astype('float32')
X_test=X_test.astype('float32')
X_train /= 255
X_test /= 255

#Change labels to categories
y_train = np_utils.to_categorical(y_train, 10)
y_test = np_utils.to_categorical(y_test, 10)
```

```python
#Create a sequential model
model = Sequential()

#Add layers
model.add(Dense(800, input_dim=784, init="normal", activation="relu"))
model.add(Dense(10, init="normal", activation="softmax"))

#compiling the model
model.compile(loss="categorical_crossentropy", optimizer="SGD", metrics=["accuracy"])

print(model.summary())
```

```python
# training
model.fit(X_train, y_train, batch_size=50, nb_epoch=125, validation_split=0.2, verbose=1)

#Estimate the quality of the model on the test set
score = model.evaluate(X_test, y_test, verbose=0)
print("Quality on the test set is: %.2f%%" % (score[1]*100))
```

```python
#Saving the architecture of NN to JSON-file

#Generate the description of the model in JSON-format
model_json = model.to_json()

#Save the model into the file
json_file = open("mnist_model.json", "w")
json_file.write(model_json)
json_file.close()

#Saving weights of the model
model.save_weights("mnist_model.h5")
```
